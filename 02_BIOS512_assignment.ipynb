Make a histogram or KDE plot that shows the distribution of body mass in each year. *Use a different theme than the default.*

```r
#Basic Histogram, no titles or labels 

options(repr.plot.width=5, repr.plot.height=7)

p = ggplot(penguins, (aes(x = body_mass_g, color = year)))
p = p + geom_histogram(bins = 20, color = "thistle3", na.rm = TRUE) 
p = p + facet_wrap(~year, ncol = 1)

# changing default theme
p = p + theme_dark()

p
```

```r
library("scales")
```

```r
# Better Histogram
# with comma formatting as well as title and informative axis labels 

# previous code from basic histogram
p = ggplot(penguins, (aes(x = body_mass_g )))
p = p + geom_histogram(bins = 20,color ='slategray2',na.rm = TRUE) 
p = p + facet_wrap(~year,ncol = 1)
p = p + theme_dark()

# adding in some commas 
p = p + scale_x_continuous(label=comma)

# adding in some titles and axis labels 
p = p + labs(title = "Distribution of Penguins' Body Mass ", 
             subtitle = 'By Year (2007-2009)',
             caption = "Made from Palmerpenguins Dataset",
             x = "Body Mass (Grams)",
             y = "Number of Penguins")
p
```

Is there a difference between the body mass of male and female penguis? Use a scatter plot with jitter to compare the masses of male and female penguins. **Facet by `species` to see compare within each species-group.**

```r
# Creating a scatterplot to compare masses between species and sex

options(repr.plot.width=7, repr.plot.height=7)

# code for scatterplot
p = ggplot(penguins, aes(x = body_mass_g, 
                         y = sex))

# adding in jitter
p = p + geom_point (color = "slategray2", 
                    position = position_jitter(h = 0.25, w = 0), 
                    alpha = 0.5,
                    size = 1,
                    na.rm = TRUE)
# faceting by species
p = p + facet_wrap(~species,ncol = 1)




# adding in some of the previous histogram's titles and axis labels as well as commas and keeping same theme
p = p + scale_x_continuous(label=comma)
p = p + labs(title = "Distribution of Penguins' Body Mass by Sex and Species ", 
             subtitle = 'By Year (2007-2009)',
             caption = "Made from Palmerpenguins Dataset",
             x = "Body Mass (Grams)",
             y = "Sex")
p = p + theme_dark()

p
```

It seems like there is a difference between male and female penguins. Males across all species tend to weigh more.
